# **Метод машинного обучения Random Forest**

На предыдущих занятиях мы с вами познакомились с идеей построения решающих бинарных деревьев для задач классификации и регрессии. Однако на практике в чистом виде их почти не используют. Но активно применяют в ансамблевых методах, которые включают в себя два весьма эффективных подхода: бэггинг и бустинг. Именно о бэггинге с использованием деревьев и пойдет речь на этом занятии.

- **Ансамбль в машинном обучении** – это метод, который объединяет прогнозы нескольких моделей для получения более точного и стабильного прогноза.

Основная идея ансамблей состоит в том, чтобы комбинировать прогнозы нескольких базовых моделей, называемых базовыми алгоритмами или участниками ансамбля. При правильном применении ансамблей их прогнозы имеют тенденцию быть лучше, чем прогнозы любого отдельного базового алгоритма.

# **1. Основные сведения**


* **Random Forest (Случайный лес)** - это алгоритм контролируемого обучения. Он имеет две разновидности - одна используется для решения задач классификации, другая - для решения задач регрессии. Это один из наиболее гибких и простых в использовании алгоритмов. Он создает деревья решений на заданных выборках данных, получает предсказания от каждого дерева и выбирает лучшее решение путем голосования. Он также является неплохим индикатором важности признаков.

* Алгоритм Random forest объединяет несколько деревьев принятия решений, в результате чего получается лес деревьев, отсюда и название `Random Forest`. В классификаторе `Random forest` чем больше деревьев в лесу, тем выше точность.

# **2. Принцип алгоритма Random Forest**


Алгоритм Random Forest можно разделить на два этапа.

На первом этапе мы случайным образом выбираем "k" признаков из общего числа `m` признаков и строим случайный лес. На первом этапе мы действуем следующим образом:-

1.	Случайным образом выбираем `k` признаков из общего числа `m` признаков, где `k < m`.
2.	Среди `k` признаков вычисляем узел `d`, используя наилучшую точку разбиения.
3.	Разделяем узел на дочерние узлы, используя наилучшее разбиение.
4.	Повторяем шаги 1-3 до достижения `l` количества узлов.
5.	Построим лес, повторив шаги 1-4 `n` раз для создания `n` количества деревьев.

image.png

На втором этапе мы используем обученный алгоритм случайного леса для предсказания.

1. Мы берем тестовые данные и применяем правила каждого случайно созданного дерева решений, чтобы предсказать результат (класс, значение). Затем сохраняем его.
2. После этого мы подсчитываем количество голосов для каждого предсказанного объекта.
3. Наконец, чтобы получить окончательный прогноз алгоритма случайного леса, мы выбираем тот класс, который получил наибольшее количество голосов (для задачи классификации) или среднее знеачение всех предсказаний (для задачи регрессии).

# **3. Достоинства и недостатки алгоритма Random Forest**



Достоинства:

- имеет высокую точность прогнозов (на большинстве задач работает лучше линейных алгоритмов);
- точность сравнима с точностью бустинга;
- практически не чувствителен к выбросам в данных из-за случайного сэмлирования выборок методом бутсрэпа;
- не чувствителен к масштабированию (и вообще к любым монотонным преобразованиям) значений признаков, связано с выбором случайных подпространств;
- не требует тщательной настройки параметров, хорошо работает «из коробки»;
- способен эффективно обрабатывать данные с большим числом признаков и классов;
- редко переобучается, на практике добавление деревьев почти всегда только улучшает композицию (до определенного, предельного уровня);
- хорошо работает с пропущенными данными;
- сохраняет хорошую точность, если большая часть данных пропущена;
- могут быть расширены до неразмеченных данных, что приводит к возможности делать кластеризацию и визуализацию данных, обнаруживать выбросы;
- легко распараллеливать и масштабировать (увеличивать число деревьев и их глубину).

Недостатки:

- в отличие от одного дерева, результаты случайного леса сложнее интерпретировать;
- алгоритм работает хуже многих линейных методов, когда в выборке очень много разреженных признаков (тексты, Bag of words);
- случайный лес не умеет экстраполировать, в отличие от той же линейной регрессии;
- алгоритм склонен к переобучению на некоторых задачах, особенно на сильно зашумленных данных;
- для данных, включающих категориальные переменные с различным количеством уровней, случайные леса предвзяты в пользу признаков с большим количеством уровней: когда у признака много уровней, дерево будет сильнее подстраиваться именно под эти признаки, так как на них можно получить более высокое значение оптимизируемого функционала (информационный выигрыш);
- больший размер получающихся моделей. Требуется O(N∙K) памяти для хранения модели, где K – число деревьев.

# **4. Выбор признаков с помощью Random Forests**


Алгоритм выбора признаков с помощью Random Forests основан на оценке важности признаков, которая вычисляется на основе способности признака уменьшить неопределенность (например, энтропию) в каждом решающем дереве. Общая идея заключается в том, что если признак имеет большую способность уменьшать неопределенность и улучшать качество предсказания, то он будет считаться значимым.

Процесс выбора признаков с помощью Random Forests обычно включает следующие шаги:

1. Построение Random Forests: на обучающей выборке строится ансамбль решающих деревьев.

2. Оценка важности признаков: для каждого признака вычисляется его важность на основе способности уменьшать неопределенность в каждом дереве.

3. Ранжирование признаков: признаки ранжируются по их важности. Признаки с наибольшей важностью считаются наиболее значимыми.

4. Отбор признаков: наиболее значимые признаки выбираются для дальнейшего использования в модели машинного обучения.

Выбор признаков с помощью Random Forests имеет ряд преимуществ. Во-первых, этот метод учитывает взаимодействие между признаками, что позволяет более точно определить их значимость. Во-вторых, Random Forests способны обрабатывать данные с большим числом признаков и преодолевать проблему мультиколлинеарности (взаимозависимости признаков). В-третьих, Random Forests могут дать оценку важности для каждого признака и показать, какой из них оказывает наибольшее влияние на результат.

> Выбор признаков с помощью Random Forests является эффективным методом отбора наиболее значимых признаков в задачах машинного обучения, который позволяет построить более точные модели и улучшить качество предсказаний.


# **5. Разница между случайными лесами и деревьями решений**



1. Построение модели: Дерево решений строится путем разделения признаков на основе определенных условий и нахождения наиболее информативного разделения. Случайный лес, с другой стороны, строится путем совокупности нескольких деревьев решений, каждое из которых обучается на подмножестве обучающей выборки и случайном подмножестве признаков.

2. Ансамбль: Случайный лес представляет собой ансамбль деревьев решений, где каждое дерево голосует за предсказание класса. Дерево решений, с другой стороны, представляет собой отдельную модель, предсказывающую класс или значение целевой переменной.

3. Стабильность и переобучение: Деревья решений могут быть склонны к переобучению, поскольку они могут идеально подстроиться под тренировочные данные. Случайный лес обладает лучшей стабильностью и устойчивостью к переобучению, благодаря использованию нескольких деревьев и случайных подвыборок данных.


4. Скорость работы: Деревья решений обучаются быстрее, чем случайный лес, поскольку в случайном лесу требуется построение и обучение нескольких деревьев. Однако, при предсказании, скорость работы случайного леса может быть выше из-за параллельной обработки деревьев внутри ансамбля.

**Случайный лес обычно обеспечивает лучшую точность предсказания и устойчивость к переобучению, но требует большего времени для обучения и подбора гиперпараметров. Деревья решений более просты в понимании и быстрее обучаются, но могут быть менее точными и более склонными к переобучению.**

# **12. Модель классификатора Random Forest с параметрами по умолчанию**

**RandomForestClassifier** - это классификатор, реализующий метод случайного леса. Он создает ансамбль решающих деревьев на основе случайной выборки признаков и случайной выборки объектов. Классификатор прогнозирует класс объекта, основываясь на голосовании решений деревьев.

Значения параметров классификатора RandomForestClassifier:

1. n_estimators - количество деревьев в случайном лесу (по умолчанию 100).
2. criterion - функция для измерения качества разбиения (по умолчанию "gini"). Возможные значения: "gini" и "entropy".
3. max_depth - максимальная глубина деревьев (по умолчанию None). Если None, то узлы будут расширяться до тех пор, пока все листы не станут однородными.
4. min_samples_split - минимальное количество образцов, необходимое для разделения внутреннего узла (по умолчанию 2).
5. min_samples_leaf - минимальное количество образцов, необходимое для быть листом (по умолчанию 1).
6. max_features - количество признаков, которые будут использоваться при делении (по умолчанию "auto"). Возможные значения: "auto", "sqrt", "log2" или целое число.
7. bootstrap - флаг, указывающий, должны ли образцы использоваться с повторениями при построении деревьев (по умолчанию True).
8. random_state - начальное состояние генератора случайных чисел (по умолчанию None).
9. n_jobs - количество параллельных задач для выполнения (по умолчанию None).

Для использования RandomForestClassifier необходимо создать экземпляр классификатора, указать необходимые параметры (если необходимо), а затем обучить модель на тренировочных данных с использованием метода fit(). После обучения, можно использовать методы predict() и predict_proba() для прогнозирования метки класса и вероятности соответствия классам соответственно.

**Инициализируем и обучаем классификатор Random Forest при использовании 10 деревьев решений**

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=10, random_state=0)
rfc.fit(X_train, y_train)
y_pred = rfc.predict(X_test)
```

Произведем оценку точности:

```python
from sklearn.metrics import accuracy_score

print('Оценка точности модели при использовании 10 деревьев решений : {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

Теперь увеличим количество деревьев решений и посмотрим, как это повлияет на точность.

**Обратимся к атрибуту "feature_importances_" у объекта-классификатора "clf"**

Атрибут "feature_importances_" содержит информацию о важности признаков для обученной модели классификации или регрессии.

В случае классификации, значение "feature_importances_" показывает, насколько каждый признак внес вклад в прогнозы классификации. Чем выше значение, тем важнее признак.

В случае регрессии, значение "feature_importances_" показывает, насколько каждый признак внес вклад в предсказания регрессии. Чем выше значение, тем важнее признак.

Обычно, значения "feature_importances_" представлены в виде массива или списка, где каждый элемент соответствует важности определенного признака.

# **17. Матрица ошибок (Confusion matrix)**



Confusion matrix (Матрица ошибок) - это инструмент, который используется для оценки производительности моделей машинного обучения в задачах классификации. Он представляет собой таблицу, в которой сравниваются фактические значения и спрогнозированные значения модели. Confusion matrix может выглядеть следующим образом:

                     Predicted Positive | Predicted Negative
    Actual Positive |    True Positives    |   False Negatives
    Actual Negative |   False Positives   |   True Negatives

1. True Positives (TP) - это количество верно предсказанных положительных классов моделью. Они отражают количество объектов, для которых модель правильно предсказала, что они принадлежат положительному классу.

2. True Negatives (TN) - это количество верно предсказанных отрицательных классов моделью. Они отражают количество объектов, для которых модель правильно предсказала, что они принадлежат отрицательному классу.

3. False Positives (FP) - это количество неверно предсказанных положительных классов моделью. Они отображают количество объектов, для которых модель неправильно предсказала, что они принадлежат положительному классу.

4. False Negatives (FN) - это количество неверно предсказанных отрицательных классов моделью. Они отражают количество объектов, для которых модель неправильно предсказала, что они принадлежат отрицательному классу.

True Positives и True Negatives служат как показатели точности модели в правильном определении классов. False Positives и False Negatives предоставляют информацию о неверных классификациях модели.

Confusion matrix также позволяет вычислять другие метрики, такие как точность (precision), полнота (recall) и F1-мера. Они часто используются в задачах классификации для оценки производительности модели и подстройки их параметров.

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt=".0f", cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

# **18. Отчет о классификации**

Отчет о классификации предоставляет информацию о точности, отзыве, F1-мере и количестве вхождений каждой метки в y_true для каждого класса. Он также включает в себя общую точность, среднее значение точности, среднее значение отзыва и среднее значение F1-меры для всех классов.

Отчет о классификации выглядит следующим образом:

             precision    recall  f1-score   support

      class 0       0.67      0.80      0.73        10
      class 1       0.83      0.71      0.77        14
      class 2       0.55      0.73      0.63        15

    micro avg       0.70      0.75      0.72        39
    macro avg       0.68      0.75      0.71        39
    weighted avg       0.71      0.75      0.72        39

* precision - это отношение числа верно предсказанных положительных результатов (true positives) к числу всех положительных результатов, она характеризует точность прогнозов модели.

* recall (отзыв) - это отношение числа верно предсказанных положительных результатов (true positives) к общему числу истинных положительных результатов, она характеризует способность модели правильно опознать положительные результаты.

* f1-score (F1-мера) - это среднее гармоническое от точности (precision) и полноты (recall). F1-мера позволяет учесть и точность, и отзыв модели.

* support - это количество вхождений каждой метки в y_true.

* micro avg - это среднее значение для всех классов по precision, recall и f1-score. Оно учитывает вклад каждого класса в вычисление общей меры.

* macro avg - это среднее значение для всех классов, независимо от размера каждого класса. Оно не учитывает дисбаланс классов.

* weighted avg - это среднее значение для всех классов, взвешенное по числу образцов в каждом классе. Оно учитывает дисбаланс классов.

# 19. Подбор гиперпараметров Random Forest

## I. Randomized Search Cross Validation

RandomizedSearchCV - это метод случайного перебора гиперпараметров, который помогает найти наилучшие гиперпараметры для модели машинного обучения. Этот метод является альтернативой GridSearchCV, который перебирает все возможные комбинации гиперпараметров.

С помощью RandomizedSearchCV можно оптимизировать выбор гиперпараметров, чтобы найти наилучшую комбинацию, уменьшая время обучения и снижая риск переобучения. Этот метод подходит для моделей с большим количеством гиперпараметров, которые не могут быть просто перебраны с помощью GridSearchCV.

Параметры RandomizedSearchCV включают:

1. estimator: Модель, которую необходимо настроить на данных.

2. param_distributions: Словарь или список словарей, содержащий параметры, которые нужно оценить. Значения параметров задаются в виде распределений, из которых будут случайным образом выбираться значения параметров для каждого запуска.

3. n_iter: Количество итераций поиска случайных комбинаций параметров.

4. scoring: Функция оценки, используемая для выбора оптимальной комбинации параметров. Может быть строкой, определяющей предварительно определенную метрику или пользовательской функцией оценки.

5. cv: Количество перекрестных проверок, используемых при оценке модели. Значение по умолчанию - 5.

6. random_state: Используется для воспроизводимости результатов случайной генерации параметров.

7. n_jobs: Количество параллельных заданий, которые могут выполняться одновременно. Может быть целым числом или -1, чтобы использовать все доступные ядра процессора.

8. verbose: Уровень подробности вывода информации о процессе обучения и оценки модели.

9. refit: Булево значение, указывающее, следует ли обучить модель с использованием найденных оптимальных параметров.

10. return_train_score: Булево значение, указывающее, должны ли возвращаться результаты оценки для обучающего набора данных.

11. error_score: Стратегия обработки ошибок, возникающих во время оценки модели. Может принимать значение 'raise', чтобы вызывать исключения, или числовое значение, которое будет назначено оценочным значениям модели, вызвавшей ошибку.

12. pre_dispatch: Количество заданий, которые необходимо предварительно отправить заданиям в очереди.

* RandomizedSearchCV возвращает модель с найденными оптимальными гиперпараметрами, которую можно использовать для предсказания новых данных.

> RandomizedSearchCV - это мощный инструмент для оптимизации выбора гиперпараметров, который помогает найти наилучшую модель машинного обучения. Он позволяет эффективно исследовать пространство гиперпараметров, ускоряя процесс обучения моделей и улучшая их качество.

```python
from sklearn.model_selection import RandomizedSearchCV

n_estimators = [int(x) for x in np.linspace(start=100, stop=500, num=10)]
max_depth = [int(x) for x in np.linspace(10, 500, num=5)]
max_depth.append(None)
max_leaf_nodes = [int(i) for i in range(1, 100)]

# Определение пространства гиперпараметров
random_grid = {
    'n_estimators': n_estimators,
    'max_depth': max_depth,
    'max_leaf_nodes': max_leaf_nodes,
}

# Определение модели классификатора
rf_clf = RandomForestClassifier(random_state=0)

# Определение объекта RandomizedSearchCV
rf_cv = RandomizedSearchCV(
    estimator=rf_clf,
    scoring='accuracy',
    param_distributions=random_grid,
    n_iter=300,
    cv=5,
    verbose=1,
    random_state=42,
    n_jobs=-1
)

# Обучение модели с перебором гиперпараметров
rf_cv.fit(X_train, y_train)

# Получение оптимальных гиперпараметров
rf_best_params = rf_cv.best_params_
print(f"Best paramters: {rf_best_params})")

# Использование оптимальных гиперпараметров в классификаторе
rf_clf = RandomForestClassifier(**rf_best_params)
rf_clf.fit(X_train, y_train)

y_pred = rf_clf.predict(X_test)

print('Оценка точности модели при подборе параметров с помощью RandomizedSearchCV : {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

```python
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt=".0f", cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
print(classification_report(y_test, y_pred))
```

`RandomizedSearchCV` позволил нам сузить диапазон для каждого гиперпараметра. Теперь, когда мы знаем, где сосредоточить поиск, мы можем явно указать все комбинации параметров, которые необходимо попробовать. Для этого используется метод `GridSearchCV`, который вместо случайной выборки из распределения оценивает все заданные нами комбинации.

## II. Grid Search Cross Validation

GridSearchCV (Grid Search Cross Validation) - это инструмент в библиотеке scikit-learn, предназначенный для перебора всех возможных комбинаций гиперпараметров модели с использованием скользящего контроля. Этот метод позволяет систематически настраивать гиперпараметры для получения наилучших результатов модели.

Параметры GridSearchCV:
1. estimator: объект модели, для которой требуется настроить гиперпараметры. Модель должна соответствовать интерфейсу оценщика scikit-learn.
2. param_grid: словарь или список словарей с параметрами, которые будут перебираться. Ключами являются имена параметров модели, а значениями - списки значений, которые должны быть проверены.
3. scoring: опциональный параметр, позволяющий задать метрику оценки качества модели. Если параметр не указан, используется значение None, и модель оценивается на основе метода score модели по умолчанию.
4. cv: опциональный параметр, позволяющий задать стратегию для скользящего контроля. По умолчанию используется стандартный 5-кратный перекрестный контроль.
5. refit: опциональный параметр, определяющий, следует ли переобучить модель на найденных оптимальных параметрах после завершения GridSearchCV. Значение по умолчанию True.
6. verbose: опциональный параметр, определяющий уровень детализации вывода. Чем больше значение, тем более подробный вывод. При значении 0 вывод не осуществляется, а при значении 10 - выводится информация о каждом этапе.
7. n_jobs: опциональный параметр, позволяющий задать количество параллельных заданий для выполнения. Значение -1 используется для задания максимального возможного количества параллельных задач.
8. return_train_score: опциональный параметр, который позволяет включить или отключить вывод оценок результатов на обучающем наборе данных. По умолчанию значение False.

* Метод fit GridSearchCV выполняет перебор параметров и поиск по сетке в соответствии с заданным словарем или списком словарей param_grid, возвращая модель с наилучшими параметрами. Поиск осуществляется по заданной метрике оценки качества, которая может быть настроена с использованием параметра scoring.

> В результате выполнения кода будет выведена информация о наилучшей модели и ее параметрах, найденных с помощью GridSearchCV.

```python
from sklearn.model_selection import GridSearchCV

n_estimators = [int(i) for i in range(180, 200)]
max_depth = [500]
max_depth.append(None)
max_leaf_nodes = [int(i) for i in range(70, 100)]

params_grid = {
    'n_estimators': n_estimators,
    'max_depth': max_depth,
    'max_leaf_nodes': max_leaf_nodes,
}

rf_clf = RandomForestClassifier(random_state=0)

rf_cv = GridSearchCV(
    rf_clf,
    params_grid,
    scoring="accuracy",
    cv=5,
    verbose=1,
    n_jobs=-1
)


rf_cv.fit(X_train, y_train)
best_params = rf_cv.best_params_
print(f"Best parameters: {best_params}")

rf_clf = RandomForestClassifier(**best_params)
rf_clf.fit(X_train, y_train)

y_pred = rf_clf.predict(X_test)

print('Оценка точности модели при подборе параметров с помощью GridSearchCV : {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

```python
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt=".0f", cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
