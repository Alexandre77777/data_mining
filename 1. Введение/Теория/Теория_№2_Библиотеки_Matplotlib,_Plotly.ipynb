# 1. Построение графиков в Matplotlib

Matplotlib предлагает несколько способов построения графиков, каждый из которых подходит для разных целей:

**1. Подграфики (Subplots)**

- Используйте `plt.subplot()`, чтобы разделить область рисования на несколько ячеек.
- В каждой ячейке можно построить отдельный график.
- Подходит для сравнения нескольких графиков или отображения разных аспектов одного набора данных.

**Пример:**

- plt.subplots(1, 2) - создает сетку из 1 строки и 2 столбцов.
- axes[0] - ссылка на первую подплоскость (в левой части).
- axes[1] - ссылка на вторую подплоскость (в правой части).
- Функции plt.plot() и plt.scatter() добавляют графические элементы на активную подплоскость.

```
import matplotlib.pyplot as plt

# Создаем фигуру с 2 подплоскостями
fig, axes = plt.subplots(1, 2)

# Линейный график в первой ячейке
axes[0].plot([1, 2, 3], [4, 5, 6])

# Диаграмма рассеяния во второй ячейке
axes[1].scatter([1, 2, 3], [7, 8, 9])

plt.show()
```

**2. Рисование на разных фигурах (Figures)**

- Используйте `plt.figure()` для создания новой фигуры.
- Каждый график будет отображаться в отдельном окне.
- Подходит для отображения больших или сложных графиков, которые не помещаются в одном окне.

**Пример:**

- Функция plt.figure() создает новую пустую фигуру.
- Функции plt.plot() и plt.scatter() добавляют графические элементы на активную фигуру.
- Активная фигура - это та, которая была создана последней.
- Функция plt.show() отображает все активные фигуры на экране.

```
import matplotlib.pyplot as plt

# Создаем 2 отдельные фигуры
fig1 = plt.figure()
fig2 = plt.figure()

# Линейный график на первой фигуре
plt.plot([1, 2, 3], [4, 5, 6])

# Диаграмма рассеяния на второй фигуре
plt.scatter([1, 2, 3], [7, 8, 9])

plt.show()
```

**3. Вложенные оси (Twin Axes)**

- Используйте `plt.twinx()` или `plt.twiny()` для создания второй оси на том же графике.
- Обе оси будут иметь одинаковые координаты X или Y.
- Подходит для отображения данных с разными масштабами или единицами измерения.

**Пример:**

- ax.twinx() - создает вторую ось, делящую ось X с основной осью.
- Вторая ось используется для отображения второго набора данных (в данном случае, температуры в градусах Фаренгейта).
- Цвет линии второго набора данных изменен для визуального различения.

```
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# Линейный график с температурой в градусах Цельсия
ax.plot([1, 2, 3], [20, 25, 30], label="Температура (°C)")

# Вторая ось для отображения температуры в градусах Фаренгейта
ax2 = ax.twinx()
ax2.plot([1, 2, 3], [68, 77, 86], label="Температура (°F)", color="red")

ax.set_ylabel("Температура (°C)")
ax2.set_ylabel("Температура (°F)")

plt.legend()
plt.show()
```

**Выбор метода:**

Выбор метода разделения графиков зависит от ваших целей и типа данных.

- **Подграфики:** простой и удобный способ для сравнения нескольких графиков.
- **Рисование на разных фигурах:** подходит для больших или сложных графиков.
- **Вложенные оси:** отображает данные с разными масштабами или единицами измерения.

## 1. Графики линий и функций:

### 1. Линейный график (Line plot):

Линейный график используется для отображения зависимости одной переменной от другой.

```
import matplotlib.pyplot as plt

# Зависимость температуры от времени
x = [1, 2, 3, 4, 5]
y = [20, 22, 24, 26, 28]

# Построение линейного графика
plt.plot(x, y)

# Название графика
plt.title("Зависимость температуры от времени")

# Подписи осей
plt.xlabel("Время (час)")
plt.ylabel("Температура (°C)")

# Отображение графика
plt.show()
```





### 2. График с областями (Area plot):

График с областями используется для отображения зависимости одной переменной от другой, а также для визуализации объема данных между двумя линиями.

```
import matplotlib.pyplot as plt

# Зависимость скорости от времени
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

# Построение графика с областями
plt.plot(x, y)
plt.fill_between(x, y, color='skyblue', alpha=0.5)

# Название графика
plt.title("Зависимость скорости от времени")

# Подписи осей
plt.xlabel("Время (сек)")
plt.ylabel("Скорость (м/с)")

# Отображение графика
plt.show()
```

```
import matplotlib.pyplot as plt

# Данные
x = [1, 2, 3, 4, 5]
y1 = [20, 25, 30, 35, 40]
y2 = [15, 20, 25, 30, 35]

# Заполнение области
plt.fill_between(x, y1, y2, color='skyblue', alpha=0.5)

# Линии
plt.plot(x, y1, label="Линия 1", color="red")
plt.plot(x, y2, label="Линия 2", color="green")

# Настройка
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

plt.show()
```

3D-линейный график используется для отображения зависимости трех переменных.



```
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Зависимость температуры от времени и высоты
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]
z = [[20, 22, 24, 26, 28],
     [21, 23, 25, 27, 29],
     [22, 24, 26, 28, 30]]

# Построение 3D-линейного графика
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for i in range(len(z)):
    ax.plot(x, y, z[i])

# Название графика
plt.title("Зависимость температуры от времени и высоты")

# Подписи осей
plt.xlabel("Время (час)")
plt.ylabel("Высота (м)")

# Отображение графика
plt.show()
```


### 3. 3D-поверхностная диаграмма (3D surface plot):



3D-поверхностная диаграмма используется для отображения функции трех переменных.

```
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Функция трех переменных
def f(x, y):
    return np.sin(x) * np.cos(y)

# Построение 3D-поверхностной диаграммы
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

ax.plot_surface(X, Y, Z)

# Название графика
plt.title("3D-поверхностная диаграмма")

# Подписи осей
plt.xlabel("x")
plt.ylabel("y")

# Отображение графика
plt.show()
```

## 5. Специальные диаграммы:



### 1. Векторное поле (Vector field):



Векторное поле используется для отображения векторных данных, таких как скорость и направление ветра.

```
import matplotlib.pyplot as plt
import numpy as np

# Векторное поле скорости ветра
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

U, V = np.meshgrid(x, y)

# Вычисление скорости и направления ветра
speed = np.sqrt(U**2 + V**2)
direction = np.arctan2(V, U)

# Построение векторного поля
plt.quiver(x, y, U, V, color='blue')

# Название графика
plt.title("Векторное поле скорости ветра")

# Подписи осей
plt.xlabel("x")
plt.ylabel("y")

# Отображение графика
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

# Векторное поле ветра
def wind_field(x, y):
    u = 10 * np.cos(np.pi * y)
    v = 10 * np.sin(np.pi * x)
    return u, v

# Сетка
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)

# Вычисление значений векторного поля
u, v = wind_field(x, y)

# Отрисовка стрелок
plt.quiver(x, y, u, v)

# Настройка
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

Контурная диаграмма используется для отображения линий равного значения функции.

```
import matplotlib.pyplot as plt
import numpy as np

# Функция с двумя переменными
def f(x, y):
    return np.sin(x) * np.cos(y)

# Построение контурной диаграммы
fig, ax = plt.subplots()

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

ax.contour(X, Y, Z, levels=10)

# Название графика
plt.title("Контурная диаграмма")

# Подписи осей
plt.xlabel("x")
plt.ylabel("y")

# Отображение графика
plt.show()
```

## 6. Гибридные диаграммы:



### 1. Комбинированная диаграмма (Combination plot):



Комбинированная диаграмма используется для объединения нескольких типов диаграмм в одном,
что позволяет одновременно сравнивать разные типы данных.

## Моделирование траектории снаряда на Python

### Представьте себе, что мы запускаем снаряд:

- В начале снаряд находится в покое в определенной точке. В момент запуска ему сообщается начальная скорость, под действием которой он начинает двигаться. Сила тяжести притягивает снаряд к Земле, изменяя его траекторию.

- Снаряд движется по параболической траектории, поднимаясь все выше и затем падая на Землю. Высота подъема снаряда зависит от его начальной скорости. Чем больше начальная скорость, тем выше поднимется снаряд.

- Дальность полета снаряда также зависит от его начальной скорости и угла запуска. На его движение влияет сопротивление воздуха, которое замедляет его движение.

- В реальных условиях траектория снаряда отличается от идеальной параболы из-за сопротивления воздуха.

- Мы можем смоделировать движение снаряда. Для этого нужно задать начальные условия: координаты снаряда, скорость, угол запуска. Также нужно учесть силу тяжести, сопротивление воздуха.

- В результате мы получим траекторию движения снаряда, его скорость в каждый момент времени. Моделирование движения снаряда позволяет изучить влияние различных факторов на его движение, оптимизировать параметры для достижения заданной цели.

```
import numpy as np
import matplotlib.pyplot as plt

# Параметры
m = 1  # Масса
g = 9.81  # Ускорение свободного падения
```

```
# Сила тяжести
def F(x, y, z, vx, vy, vz):
    return np.array([0, -m * g, 0])
```

```
# Вычисление траектории
def trajectory(x0, y0, z0, vx0, vy0, vz0, dt, t_max):
    x = np.zeros(int(t_max / dt) + 1)
    y = np.zeros(int(t_max / dt) + 1)
    z = np.zeros(int(t_max / dt) + 1)
    vx = np.zeros(int(t_max / dt) + 1)
    vy = np.zeros(int(t_max / dt) + 1)
    vz = np.zeros(int(t_max / dt) + 1)

    x[0] = x0
    y[0] = y0
    z[0] = z0
    vx[0] = vx0
    vy[0] = vy0
    vz[0] = vz0

    for i in range(1, int(t_max / dt) + 1):
        # Сила
        f = F(x[i - 1], y[i - 1], z[i - 1], vx[i - 1], vy[i - 1], vz[i - 1])

        # Положение
        x[i] = x[i - 1] + vx[i - 1] * dt
        y[i] = y[i - 1] + vy[i - 1] * dt
        z[i] = z[i - 1] + vz[i - 1] * dt

        # Скорость
        vx[i] = vx[i - 1] + f[0] * dt / m
        vy[i] = vy[i - 1] + f[1] * dt / m
        vz[i] = vz[i - 1] + f[2] * dt / m

    return x, y, z
```

```
# Параметры траектории
x0 = 0  # Начальная координата X
y0 = 0  # Начальная координата Y
z0 = 0  # Начальная координата Z
vx0 = 10  # Начальная скорость X
vy0 = 10  # Начальная скорость Y
vz0 = 0  # Начальная скорость Z
dt = 0.01  # Шаг времени
t_max = 10  # Максимальное время

# Вычисление траектории
x, y, z = trajectory(x0, y0, z0, vx0, vy0, vz0, dt, t_max)

# 3D-график
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Траектория
ax.plot3D(x, y, z, color='red')

# Настройка
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

# 2. Примеры построения графиков в Plotly Express:



**1.1. Гистограмма:**



Гистограмма показывает распределение данных в виде столбцов. Она помогает визуализировать, как часто встречаются разные значения в наборе данных.



```
import plotly.express as px

df = px.data.tips()

fig = px.histogram(df, x="total_bill")
fig.show()
```

**1.2. "Ящик с усами":**



Коробчатая диаграмма (Ящик с усами) показывает медиану, квартили, выбросы и другие статистические характеристики данных. Она помогает сравнить распределение данных в разных группах.



```
fig = px.box(df, y="total_bill")
fig.show()
```

**1.3. Диаграмма рассеяния:**

Диаграмма рассеяния показывает точечное представление данных с двумя переменными. Она помогает увидеть связь между двумя переменными.

```
fig = px.scatter(df, x="total_bill", y="tip")
fig.show()
```





**1.4. График плотности:**

График плотности показывает распределение вероятностей данных. Он помогает увидеть, как часто встречаются разные значения в наборе данных. Педставляет собой контурную диаграмму плотности распределения.

```
fig = px.density_contour(df, x="total_bill")
fig.show()
```



**2.1. Линейный график:**


Линейный график показывает изменения значения одной переменной в зависимости от другой. Он помогает увидеть тренды в данных.

```
fig = px.line(df, x="total_bill", y="tip")
fig.show()
```






**2.2. График с областями:**

График с областями показывает область, в которой находятся данные. Он помогает увидеть, насколько изменчивы данные.

```
fig = px.area(df, x="total_bill", y="tip")
fig.show()
```





**2.3. Столбчатый график:**

Столбчатый график показывает сравнение категориальных данных. Он помогает увидеть, как разные категории сравниваются друг с другом.


```
fig = px.bar(df, x="day", y="total_bill")
fig.show()
```




**2.4. Круговая диаграмма:**

Круговая диаграмма показывает пропорции категориальных данных. Она помогает увидеть, как разные категории составляют целое.

```
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]
tips = [10, 15, 12, 8, 20]

# Создать pie-диаграмму
fig = px.pie(values=tips, names=days)

# Отобразить диаграмму
fig.show()
```



**3.1. 3D-точечная диаграмма:**



3D-точечная диаграмма показывает точечное представление данных с тремя переменными. Она помогает увидеть связь между тремя переменными.



```
total_bill = np.array([10, 15, 12, 8, 20])
tip = np.array([10, 15, 12, 8, 20])
day = np.array(["Mon", "Tue", "Wed", "Thu", "Fri"])

fig = px.scatter_3d(x=total_bill, y=tip, z=day)

fig.show()
```



**4.1. Матрица рассеяния:**

Матрица рассеяния показывает матрицу диаграмм рассеяния для всех пар переменных (`data`) в наборе данных. Она помогает увидеть связь между каждой парой переменных.

```
import numpy as np
import plotly.express as px

# Создаем массивы данных
data = np.random.rand(100, 4)

# Рисуем матрицу рассеяния
fig = px.scatter_matrix(data)
fig.show()
```



**5.1. Карта хорроплет:**

Карта хорроплет показывает распределение данных (`data`) по географическим регионам (`countries`). Она помогает увидеть, как данные распределены по миру.

```
import plotly.express as px

df = px.data.election()
geojson = px.data.election_geojson()

fig = px.choropleth(df, geojson=geojson, color="Bergeron",
                    locations="district", featureidkey="properties.district",
                    projection="mercator"
                   )
fig.update_geos(fitbounds="locations", visible=False)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```





**5.2. Карта точек:**



Карта точек показывает точечное представление данных (`latitudes`, `longitudes`) на карте. Она помогает увидеть, где находятся данные.



```
# Создаем второй массив данных
latitudes = np.random.uniform(-90, 90, 50)
longitudes = np.random.uniform(-180, 180, 50)

# Рисуем карту точек
fig = px.scatter_geo(lat=latitudes, lon=longitudes)
fig.show()
```
