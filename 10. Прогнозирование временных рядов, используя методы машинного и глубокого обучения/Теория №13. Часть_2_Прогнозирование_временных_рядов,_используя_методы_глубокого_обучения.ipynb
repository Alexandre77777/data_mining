## RNN

Ссылка на оригинальную статью: https://arxiv.org/abs/1406.1078

Рекуррентные нейронные сети (RNN — Recurrent Neural Networks) представляют собой класс нейронных сетей, способных обрабатывать последовательности произвольной длины за счет внутренних состояний, сохраняемых в ячейках памяти. Стандартные нейронные сети неспособны сохранять информацию о предыдущих данных во время обработки текущих, что является ограничением при работе с последовательными данными, такими как временные ряды, тексты и аудиозаписи. В отличие от них, RNN использует свою память для обработки последовательности входных данных, сохраняет информацию о предшествующем контексте и переносит эту информацию через цепочку последующих обработок.

Принцип работы RNN заключается в следующем:
1. **Состояние сети (hidden state)**: В сердце RNN находится вектор скрытого состояния, который обновляется на каждом шаге обработки входных данных. Для каждого элемента входной последовательности нейронная сеть обновляет свое скрытое состояние посредством рекуррентной функции.
   
2. **Рекуррентная функция**: Функция обновления скрытого состояния представляет собой нейронную сеть, которая принимает предыдущее скрытое состояние и текущий вход и вычисляет новое скрытое состояние. Это состояние может включать в себя информацию из всей предыдущей последовательности, что позволяет сети делать выводы на основе предшествующего контекста.

3. **Веса**: Как и в других типах нейронных сетей, RNN используют веса для управления значимостью входных данных и предыдущего состояния. Эти веса обновляются в процессе обучения сети, что позволяет модели адаптироваться к особенностям обрабатываемых данных.

4. **Выход**: На каждом шаге входной последовательности RNN может производить выход. Этот выход вычисляется на основе текущего скрытого состояния, которое суммирует информацию предыдущих шагов обработки.

5. **Обратное распространение через время (BPTT— Backpropagation Through Time)**: Специальная версия обратного распространения, используемая в RNN для обучения, связана с распространением градиента назад во времени и обновлением весов на всех шагах последовательности.

Основной недостаток RNN — это проблема затухающих и взрывающихся градиентов, когда в процессе обучения изменения весов становятся либо слишком маленькими (затухающие), либо слишком большими (взрывающиеся), что делает трудным сохранение информации через длинные временные промежутки. Для решения этой проблемы были разработаны различные архитектуры, такие как LSTM (Long Short-Term Memory) и GRU (Gated Recurrent Units), которые представляют собой усовершенствованные версии RNN.

## LSTM

Ссылка на оригинальную статью: https://arxiv.org/abs/1402.1128

LSTM, что расшифровывается как Long Short-Term Memory, представляет собой разновидность рекуррентных нейронных сетей (RNN), ориентированных на работу с последовательностями данных и устранение проблем, связанных с долгосрочными зависимостями.

Традиционные RNN страдают от проблемы исчезающего и взрывающегося градиента при обучении, которая ведет к тому, что модель неэффективно обучается на данных с длинными зависимостями. LSTM решает эту проблему через введение ячеек памяти, которые способны поддерживать информацию на произвольно долгому промежутке времени.

Ключевые элементы архитектуры LSTM:

1. **Вентили (gates) –** это компоненты LSTM-ячеек, которые регулируют состояние ячейки.
   - **Вентиль забывания (forget gate):** определяет, какая информация из прошлого состояния ячейки должна быть забыта.
   - **Входной вентиль (input gate):** решает, какая новая информация будет добавлена в состояние ячейки.
   - **Выходной вентиль (output gate):** контролирует, какая часть информации из состояния ячейки передается далее по сети.

2. **Состояние ячейки (cell state) –** это «дорожка памяти» через весь слой LSTM, информация по которой передается с минимальными изменениями, и может быть прочитана или модифицирована через вентили.

3. **Скрытое состояние (hidden state) –** вектор, который передает краткосрочную информацию к следующему временному шагу и к выходу слоя.

Принцип работы LSTM основан на последовательных обновлениях состояния ячеек и скрытых состояний, что позволяет не только различать важную информацию от неважной, но и удерживать ее на необходимый период времени. Математически эти обновления описываются серией уравнений, использующих такие операции, как поэлементное умножение (Hadamard product) и активационные функции (такие как сигмоид для вентилей и тангенс гиперболический для обновления состояния ячейки).

Обучение LSTM, как и большинства нейронных сетей, происходит с использованием алгоритма обратного распространения ошибки во времени (Backpropagation Through Time, BPTT). Однако, чтобы избежать проблемы исчезающего градиента, LSTM использует свои уникальные структуры вентилей для поддержания градиента на нужном уровне.

LSTM широко применяются для задач обработки естественного языка (Natural Language Processing, NLP), распознавания речи, прогнозирования временных рядов и в других областях, где данные представляют собой последовательности различной длины.

## TCN

Ссылка на оригинальную статью: https://arxiv.org/abs/1803.01271

Модель Temporal Convolutional Network (TCN), или временная свёрточная сеть, представляет собой архитектуру глубокого обучения, оптимизированную для работы с временными рядами и последовательностями. TCN сочетает в себе преимущества свёрточных нейронных сетей (CNN) для обработки изображений и рекуррентных нейронных сетей (RNN) для анализа временных рядов, но без некоторых их недостатков, в частности проблемы исчезающего градиента, характерной для RNN.

**Основные характеристики TCN:**

1. **Сверточные слои**: В отличие от традиционных CNN, использующих свертки для обработки пространственных структур, TCN применяют одномерные свертки для анализа временных последовательностей.

2. **Повышенная рецептивная область**: Для учёта долгосрочных зависимостей в последовательностях TCN используют концепцию разреженных сверток (dilated convolutions). Это позволяет с каждым сверточным слоем экспоненциально увеличивать рецептивное поле без потери разрешения и увеличения количества параметров.

3. **Остаточные блоки (Residual Blocks)**: Для борьбы с проблемой исчезающего или взрывающегося градиента архитектура TCN включает остаточные связи, позволяющие градиентам течь более эффективно во время процесса обучения.

4. **Принцип причинности (Causality)**: Временная свёрточная сеть обеспечивает строгий порядок в данных — выход в момент времени `t` может зависеть только от элементов в моменты времени `t` и ранее. Это достигается путем применения padding таким образом, чтобы сверточные фильтры не "заглядывали" в будущее.

**Принцип работы TCN:**

- **Обработка данных**: Входными данными для TCN являются временные ряды. Формат подачи данных похож на подходы, используемые в RNN, но, в отличие от RNN, TCN обрабатывает все шаги времени одновременно, что обеспечивает высокую параллельность обработки.

- **Сверточные слои**: Данные проходят через одномерные сверточные слои, где ширина фильтров определяет, сколько временных шагов они покрывают. Для учета долгосрочных зависимостей используются разреженные свертки.

- **Остаточные блоки**: Вход каждого сверточного блока прибавляется к его выходу, что содействует более быстрому обучению и предотвращает затухание или взрыв градиента для глубоких сетей.

- **Нормализация**: Нормализация по слоям может применяться для стабилизации и ускорения обучения.

- **Обратное распространение ошибки**: Значения потерь распространяются назад через сеть для её обучения, подобно обратному распространению в других типах нейронных сетей.

TCN может быть использована в различных задачах, связанных с временными рядами, включая прогнозирование, обнаружение аномалий и классификацию последовательностей, предлагая превосходное производительное решение благодаря своей способности к быстрому прогнозированию и обработке.

## DilatedRNN

Ссылка на оригинальную статью: https://arxiv.org/abs/1710.02224

Dilated Recurrent Neural Network (DilatedRNN) – это архитектура нейронной сети, которая улучшает обработку последовательностей во времени, расширяя стандартный RNN с помощью дилатационных (расширенных) связей. Данный подход повышает способность сети обрабатывать информацию на разных временных масштабах и помогает уменьшить проблему исчезающего градиента, которая традиционно присуща RNN.

**Принцип работы DilatedRNN:**

1. **Дилатация (расширение):** В DilatedRNN добавляются так называемые дилатационные связи. Это значит, что нейроны на более глубоких слоях сети соединяются не с каждым последовательным входным элементом, а пропускают некоторое количество элементов в последовательности. Например, с коэффициентом дилатации 2, нейрон второго слоя будет получать информацию только от каждого второго нейрона первого слоя, что позволяет охватывать более длинные временные интервалы без увеличения числа параметров или размера сети.

2. **Слои с увеличивающейся дилатацией:** DilatedRNN часто строят таким образом, что каждый новый слой имеет увеличенный коэффициент дилатации в геометрической прогрессии. Это подобно приему в сверточных нейронных сетях, где увеличение размера поля восприятия помогает улавливать признаки более высокого уровня абстракции. В DilatedRNN подобная архитектура помогает модели "видеть" дальше во времени с каждым новым слоем.

3. **Параллелизм во времени:** В отличие от традиционных RNN, где расчеты должны последовательно пройти через все моменты времени, DilatedRNN позволяет выполнять некоторые расчеты параллельно. Это обусловлено тем, что данные, поступающие на вход различных нейронов, не зависят непосредственно друг от друга из-за пропусков, вводимых дилатацией.

4. **Уменьшение проблемы исчезающего градиента:** Поскольку дилатированные связи позволяют сигналам и градиентам обходить некоторые уровни и тем самым передаваться на большие расстояния по времени с меньшими потерями, проблема затухания градиента может быть уменьшена, что позволяет лучше учиться на долгих последовательностях данных.

5. **Применение:** DilatedRNN показывают высокую эффективность в задачах, где необходимо учитывать контекст на различных временных масштабах, таких как анализ временных рядов, обработка звука, музыки или языковые моделирования.

Архитектура DilatedRNN снимает некоторые ограничения, связанные с длиной зависимостей во времени, характерные для классических RNN, и позволяет более эффективно моделировать сложные временные отношения в данных без чрезмерного увеличения количества параметров.

## N-BEATS

Ссылка на оригинальную статью: https://arxiv.org/abs/1905.10437

N-BEATS (Neural Basis Expansion Analysis for Time Series) — это архитектура нейронной сети, разработанная для прогнозирования временных рядов с высокой точностью. Эта модель была представлена в статье Oreshkin et al. и отличается от классических рекуррентных сетей, таких как LSTM и GRU, тем что она использует полносвязные слои и работает с предположением, что любой временной ряд можно декомпозировать на базисные функции.

Ключевым аспектом N-BEATS является использование двух типов блоков: трендовых и сезонных. Трендовый блок научается улавливать долгосрочные зависимости в данных, тогда как сезонный блок фокусируется на периодичных паттернах.

В архитектуре модели используются стеки блоков, каждый из которых генерирует прогноз и уточняет остатки, или ошибки, предыдущего блока. На практике это означает, что сначала используется блок для поимки трендовой компоненты временного ряда, после чего остатки, не объясненные трендом, передаются следующему блоку, который может уловить сезонность и так далее. Этот процесс похож на иерархическую регрессию, где каждый следующий уровень аппроксимирует ошибку предыдущего.

Каждый блок N-BEATS включает в себя следующие компоненты:
1. **Входной слой**, который обрабатывает входные данные временного ряда.
2. **Слой расширения (Expansion layer)**, где входной сигнал проецируется на высокоразмерное пространство для выявления потенциально сложных закономерностей.
3. **Слой выбора базисной функции (Basis selection layer)**, который трансформирует высокоразмерные данные обратно в пространство прогноза временного ряда. Здесь используются параметризованные базисные функции для моделирования трендов и сезонности.
4. **Прогнозирующий блок (Forecasting block)**, генерирующий конечный прогноз временного ряда.

Эти блоки используют прямые связи, что позволяет обеспечить быстрое обучение и защищает модель от проблемы исчезающего или взрывающегося градиента, которая встречается в рекуррентных архитектурах.

Применение N-BEATS удобно в сценариях, где требуются точные прогнозы без предварительного знания об особенностях временных рядов, так как она способна фактически "вычленить" основные характеристики данных самостоятельно.

Данная модель продемонстрировала выдающиеся результаты на разнообразных временных рядах и стала популярным инструментом в области машинного обучения для прогнозирования временных рядов благодаря своей способности к генерализации и точности предсказаний.

N-BEATS (Neural Basis Expansion Analysis for Time Series Forecasting) представляет собой архитектуру нейронной сети, предназначенную для прогнозирования временных рядов. Эта модель была представлена в статье, написанной Oreshkin и соавторами, и выделяется своей способностью к точному и интерпретируемому моделированию временных рядов благодаря новаторскому подходу абстракции от типичных рекуррентных и сверточных сетевых структур.

**Архитектура N-BEATS:**

N-BEATS построена на блоках, каждый из которых снабжен двумя полносвязными сетями (feed-forward neural networks), которые называются передаточными (backcast) и прогнозирующими (forecast) сетями. Эти сети моделируют соответственно прошлое и будущее временного ряда.

**Основные компоненты:**

1. **Блоки:** Модель состоит из нескольких идентичных блоков, последовательно уложенных друг на друга. Каждый блок сам по себе содержит две части: одна для восстановления входных данных (backcast) и одна для предсказания (forecast).
   
2. **Stack architecture:** Каждый последующий блок оперирует остатками от предыдущего блока, что позволяет модели учиться на ошибках предыдущих слоев и прогрессивно сочленять окончательный прогноз.
   
3. **Basis expansion:** Важной чертой N-BEATS является использование базисного расширения (basis expansion). Это означает, что модель рассматривает прогноз как сумму функций, которые определяются весами нейронной сети. Модель может быть двух видов - обобщенная и интерпретируемая. В обобщенной версии используются полинымы, а в интерпретируемой версии - предопределенные паттерны временных рядов, такие как тренды или сезонность.

4. **Double Residual Stacking:** В каждом блоке используется метод двойного остатка (double residual stacking). Сначала идет вычитание - backcast, которое удаляет уже моделируемую часть временного ряда, затем добавление - forecast, которое агрегирует прогнозируемые будущие значения.

**Принцип работы:**

Этапы работы модели N-BEATS:

1. Исходный временной ряд подается на вход первому блоку N-BEATS.
   
2. Каждый блок порождает два результата: backcast, который оценивает и вычитает соответствующую составляющую из входного временного ряда, и forecast, который формирует часть окончательного прогноза.
   
3. Остаток временного ряда, т.е. часть не захваченная моделированием блока, подается на вход следующему блоку.
   
4. Процедура повторяется для каждого блока, пока все блоки не пройдут через процесс. Прогнозы каждого блока агрегируются, формируя итоговый прогноз временного ряда.

N-BEATS показала высокую точность в разнообразных задачах прогнозирования временных рядов и отличается от других моделей своей интерпретируемостью и широкими возможностями адаптации под конкретные виды задач.

## N-HiTS

Ссылка на оригинальную статью: https://arxiv.org/abs/1905.10437

N-HiTS (N-Hierarchical Time-Series) — это архитектура нейронной сети, разработанная для прогнозирования временных рядов, представляющая собой разновидность модели N-BEATS (Neural Basis Expansion Analysis for Time Series). Однако, в отличие от N-BEATS, которая использует обратную связь (feedback loop) для улучшения предсказания, N-HiTS вводит иерархическую структуру с несколькими горизонтами прогнозирования, что позволяет модели эффективно справляться с разнообразными шаблонами в данных и разными временными масштабами.

Архитектура N-HiTS состоит из нескольких ключевых компонентов:

1. **Иерархические блоки**: Модель включает в себя ряд уровней иерархии, каждый из которых адаптируется к определенному временному разрешению. Это позволяет модели одновременно работать с долгосрочными и краткосрочными зависимостями в данных.

2. **Внимание к разрешению**: N-HiTS применяет механизмы внимания, чтобы настроиться на полезные сигналы при разных временных интервалах.

3. **Блоки прогнозирования**: Внутри каждого иерархического блока располагаются блоки, выполняющие предсказание, сходные с блоками в N-BEATS, но адаптированные для работы на разных уровнях разрешения.

4. **Повторное использование параметров**: Блоки могут повторно использовать параметры для разных временных горизонтов, что уменьшает риск переобучения и упрощает обучение модели.

5. **Уточнение прогнозов**: Прогнозы с различных иерархических уровней объединяются для уточнения общего предсказания.

Как и N-BEATS, N-HiTS выделяется тем, что не опирается на предыдущие прогнозы для генерации будущих значений, благодаря чему снижается вероятность накопления ошибок прогноза.

Принцип работы заключается в следующем:

- Временной ряд вначале обрабатывается на множестве иерархических уровней, каждый из которых выделяет особенности конкретного временного масштаба.
- Сигналы с разных уровней затем поступают в блоки прогнозирования, где прогнозы формируются независимо.
- После создания прогноза, модель использует механизмы взвешивания для объединения предсказаний из разных иерархических уровней.
- Объединённый прогноз дает конечное предсказание временного ряда.

Преимущества N-HiTS перед другими моделями заключаются в гибкости и мощности архитектуры, способной улавливать как долгосрочные, так и краткосрочные закономерности благодаря использованию иерархических блоков. Эта модель показывает обещающие результаты на задачах прогнозирования временных рядов, где требуется учёт широкого спектра временных зависимостей.

## Transformer

Ссылка на оригинальную статью: https://arxiv.org/abs/1706.03762

Модель нейронных сетей Transformer представляет собой архитектурный подход, предложенный в 2017 году для обработки последовательностей данных, таких как тексты. Данная модель стала особенно популярной в задачах машинного перевода.

Основными компонентрами базовой модели Transformer являются механизм внимания (attention) и многоуровневая структура нейронной сети. Задача модели состоит в том, чтобы перевести или генерировать выходные последовательности, исходя из входных последовательностей.

Transformer состоит из двух основных частей: энкодера и декодер. Каждая часть состоит из нескольких слоев, называемых блоками трансформера. Каждый блок состоит из двух подслоев: механизма внимания применяемого к различным позициям входных последовательностей и полносвязного прямого прохода (feed-forward).

Механизм внимания Transformer позволяет модели фокусироваться на различных аспектах входных данных в каждом блоке. Для этого используется матричное умножение, в результате которого вычисляется весовая комбинация всех входных позиций для каждого элемента последовательности. Это позволяет учитывать контекст и отношения между различными частями последовательности, что является ключевым фактором для достижения хороших результатов в задачах машинного перевода.

Каждый блок энкодера и декодера также содержит полносвязные слои, применяемые после механизма внимания. Эти слои выполняют операции линейного преобразования и нелинейное преобразование с помощью функции активации (например, ReLU). Таким образом, блоки трансформера выполняют операции внимания и преобразования данных последовательности на различных уровнях.

В процессе обучения модели используется метод обратного распространения ошибки, который позволяет модели настраивать веса, чтобы минимизировать ошибку между предсказанными и правильными значениями. Для этого используется функция потерь, которая может различаться в зависимости от задачи.

В целом, модель нейронных сетей Transformer представляет собой эффективный подход к обработке последовательностей данных, позволяющий учитывать контекст и взаимосвязи между элементами последовательности. Это делает модель очень успешной в задачах машинного перевода и других задачах обработки естественного языка.

Принцип работы Трансформера следующий:

1. Входная последовательность преобразуется в векторы эмбеддингов, к которым добавляется позиционное кодирование.

2. Преобразованные данные поступают в блоки множественного внимания, где обучаемые веса определяют, на что модель должна обращать внимание в каждом шаге.

3. После масштабируемого произведения внимания результат проходит через полносвязную сеть.

4. Далее, данные передаются через нормализацию и residual connections к следующему слою.

5. Аналогичная структура используется в декодере с той разницей, что добавляется еще один слой внимания, который учитывает выходные данные энкодера.

6. На выходе декодера находится полносвязный слой, который преобразует векторы декодера в вероятности для следующего элемента выходной последовательности.

## Informer

Ссылка на оригинальную статью: https://arxiv.org/abs/2012.07436

Модель Informer является одной из архитектур глубокого обучения для прогнозирования временных рядов длинного диапазона. Эта нейросеть была разработана для улучшения способности захвата долговременных зависимостей в данных без потери информации, что является сложной задачей для многих существующих моделей, таких как RNN и LSTM.

**Архитектурные особенности Информера:**

1. **Проборная механика внимания (ProbSparse Self-attention):** Informer использует оптимизированный механизм внимания, где во время рассчета внимания выбираются ключевые запросы, что позволяет сократить количество вычислений и улучшить обработку длинных последовательностей.

2. **Многомерное масштабирование (Multi-scale Attention):** Функция позволяет модели генерировать представления на различных временных масштабах, обеспечивая улавливание зависимостей на разном уровне временного разрешения.

3. **Иерархическая структура (Hierarchical Structure):** Informer внедряет иерархическую структуру для дистилляции информации из данных с высокой частотой. Это обеспечивает более эффективную обработку и сокращение размерности данных.

4. **Обработка различных временных масштабов (Distilling Operation):** Informer применяет операцию дистилляции, которая консолидирует информацию с течением времени для уменьшения количества требуемых вычислений и ускорения инференции.

5. **Прогнозирование последовательностей переменной длины (Variable-length Sequence Forecasting):** Архитектура позволяет модели обрабатывать и предсказывать временные ряды переменной длины без необходимости обрезки или дополнения последовательностей.

6. **Механизм самовнимания (Self-attention Mechanism):** Как и трансформер, Informer использует ряд слоев самовнимания, которые позволяют каждому элементу последовательности независимо оценивать свою связь с другими элементами.

**Принципы работы:**

- Informer обрабатывает входные данные временного ряда с использованием структуры автоэнкодера, где энкодер изучает внутренние представления временного ряда, а декодер генерирует последующие прогнозы.

- Модель применяет слои самовнимания для выявления и кодирования информационных взаимодействий между различными моментами временного ряда.

- Informer стремится минимизировать потери предсказания за счет использования большого объема информации и прогнозирования на значительные временные горизонты с высокой точностью.

- Модель способна адаптироваться к динамическим временным рядам и изменяющимся временным масштабам за счет своей гибкости и адаптивности к данным разного размера и ритма.

Используя эти инновации, Informer показывает высокую производительность на задачах с длинными временными рядами, превосходя традиционные методы и общие архитектуры нейронных сетей в определенных условиях.

## TFT

Ссылка на оригинальную статью: https://www.sciencedirect.com/science/article/pii/S0169207021000637

Temporal Fusion Transformer (TFT) - это модель для прогнозирования временных рядов, разработанная для улучшения точности и интерпретируемости. Эта архитектура совмещает идеи из рекуррентных нейронных сетей (RNN), сверточных нейронных сетей (CNN) и трансформеров (в частности, механизмов внимания), и предназначена для обработки разнородных временных рядов с возможностью учета экзогенных факторов.

В основе модели TFT лежат следующие компоненты:

1. **Gated Residual Network (GRN)**:
   - Это ключевой элемент TFT, применяемый для обработки как временно́-зависимой, так и статической информации. В каждом GRN используется механизм "вентелей", аналогичный тому, что используется в LSTM, для управления потоком информации, обучаясь и фильтруя только релевантные входные данные.

2. **Variable Selection Networks**:
   - Эти сети определяют значимость как временно́-зависимых, так и статических сигналов с использованием механизмов внимания. Такой подход позволяет модели автоматически отбирать наиболее информативные признаки.

3. **Static Covariate Encoders**:
   - Кодировщики обрабатывают статические (не изменяющиеся со временем) переменные, чтобы улучшить контекст понимания временных рядов с помощью внимания, основанного на GRN.

4. **Temporal Process Blocks**:
   - Блоки состоят из последовательности слоев, включающих GRN и временно́-зависимые механизмы внимания, что позволяет улавливать долгосрочные зависимости во времени.

5. **Sequence-to-Sequence Self-Attention Mechanism**:
   - Архитектура TFT содержит механизмы внимания, которые помогают модели обращать внимание на разные участки входных данных, по аналогии с трансформерами, которые были первоначально разработаны для задач обработки естественного языка.

6. **Interpretable Multi-Head Attention**:
   - TFT использует механизмы множественного внимания для того, чтобы модель могла параллельно учиться на различных аспектах входных данных, с возможностью последующей интерпретации важности разных входных признаков.

7. **Quantile Output Layer**:
   - Выходной слой TFT обучается предсказывать не только средние значения, но и квантили прогнозируемого распределения, что позволяет оценивать неопределенность и подготавливать более информативные прогнозы.

Архитектура TFT направлена на задачи, где важна точность прогноза и его интерпретируемость, что делает её популярной во многих приложениях: от финансового моделирования до энергетики и управления запасами. Модель была специально разработана для работы с гетерогенными данными, содержащими как числовые, так и категориальные переменные, обладающими различными временными характеристиками.

## PatchTST (Patch Temporal Shift Transformer)

Ссылка на оригинальную статью: https://arxiv.org/pdf/2211.14730.pdf

PatchTST (Patch Temporal Shift Transformer) – это модель нейронной сети, предназначенная для анализа временных рядов и видео, которая объединяет понятия из современных трансформерных архитектур с принципами сверточных нейронных сетей. Она призвана улучшить извлечение пространственно-временных признаков за счет использования механизма сдвига, применяемого к патчам – разделенным частям входных данных. Ниже представлены ключевые аспекты, характеризующие ее принципы работы:

1. **Патчи**: В PatchTST входные данные (будь то одномерные временные ряды или многомерные видеоряды) разбиваются на небольшие, управляемые блоки, известные как патчи, прежде чем они будут поданы в модель. Это облегчает извлечение признаков и позволяет уменьшить вычислительную сложность.

2. **Сдвиг Времени**: Применяется операция сдвига времени для каждого патча, которая позволяет явно моделировать временные отношения. Сдвиги во временном измерении помогают улучшить сверточное изучение признаков, делая модель более внимательной к динамике входных данных.

3. **Трансформер**: Основой PatchTST является архитектура трансформера, оптимизированного для обработки последовательностей патчей. Трансформер использует механизмы самовнимания, которые могут захватывать сложные зависимости на любых дистанциях временной последовательности, обеспечивая динамичное взвешивание признаков на основе их взаимной важности.

4. **Обработка Информации**: Подход сдвига времени в модели PatchTST позволяет обработать больше контекста, что особенно ценно для прогнозирования и понимания сложных временных рядов. Он обеспечивает эффективный и глубокий обзор временных признаков без чрезмерной вычислительной нагрузки.

5. **Обучение и Эффективность**: PatchTST характеризуют возможности распараллеливания обработки, что позволяет значительно ускорить обучение по сравнению с традиционными рекуррентными моделями. Это делает ее особенно подходящей для работы с большими объемами данных и скоростными потоками в реальном времени.

PatchTST интегрирует пространственные и временные домены в единую обучаемую модель, что дает ей возможность улавливать более сложные пространственно-временные зависимости, чем традиционные сверточные или рекуррентные сети. Это делает ее перспективной для применения в таких областях, как анализ видео, обработка сигналов, предсказание временных рядов и многих других задачах, где необходимо учитывать динамическое изменение данных во времени.

## StemGNN (Spatiotemporal Multi-Graph Network)

Ссылка на оригинальную статью: https://arxiv.org/abs/2103.07719

StemGNN (Spatiotemporal Multi-Graph Network) является передовой моделью нейронных сетей для прогнозирования временных рядов, которые демонстрируют пространственно-временные зависимости. Модель была разработана для более эффективного анализа и моделирования данных, где пространственные связи и временные закономерности взаимосвязаны и изменяются с течением времени.

Основное отличие StemGNN от традиционных подходов состоит в том, что она характеризуется способностью одновременно обрабатывать несколько различных видов графовых структур в данных. Это особенно важно в сценариях, где взаимосвязи между точками данных могут быть многомерными и представлены в виде нескольких графов — например, географические данные, социальные сети, или транспортные системы.



Принцип работы StemGNN основан на использовании сверточных нейронных сетей для извлечения признаков из графовых данных. Основная идея состоит в том, чтобы преобразовать соседние узлы и ребра графа в векторные представления и использовать их для обновления самого узла.

Алгоритм работы StemGNN состоит из следующих шагов:

1. Инициализация входных данных: Входные данные представляются в виде графа, где узлы представляют объекты, а ребра - их взаимосвязи. Каждый узел содержит некоторую информацию о свойствах объекта.

2. Кодирование узлов и ребер: Для каждого узла и его соседей выполняется кодирование, где информация о свойствах узла и ребер конвертируется в векторное представление.

3. Сверточные слои: Затем выполняются сверточные операции, где каждый узел обновляется на основе его векторного представления и соседних узлов. Для этого используются сверточные слои, которые обрабатывают информацию о свойствах узлов и ребер и объединяют ее, чтобы получить новое представление узла.

4. Пуллинг графа: После выполнения сверточных операций может быть выполнено пуллинг графа, где узлы графа агрегируются, чтобы получить общее представление графа.

5. Выходные данные: На последнем шаге, после обработки всех узлов и выполнения пуллинга, получается финальное представление графа, которое может быть использовано для решения задачи, таких как классификация, сегментация или генерация новых данных.

Основные принципы работы StemGNN:

1. **Представление данных в виде графов**: StemGNN эффективно решает задачу когда пространственные взаимосвязи между точками данных могут быть закодированы в виде графа.

2. **Использование многографов**: Так как реальные пространственно-временные данные часто содержат множественные типы связей, StemGNN использует несколько графов для представления различных видов взаимосвязей или потоков информации между узлами.

3. **Спектральная обработка**: Архитектура применяет спектральные методы для изучения временных зависимостей, используя преобразование Фурье для преобразования временных рядов в частотное пространство, что делает возможным анализ частотных паттернов.

4. **Гибридное пространственно-временное моделирование**: StemGNN объединяет пространственное и временное моделирование для учета сложных зависимостей и динамики.

5. **Многозадачное обучение**: Модель может быть обучена на основе нескольких задач прогнозирования, сочетая прогнозы для разных переменных или временных горизонтов, что улучшает общее качество моделирования через совместное извлечение признаков.

6. **Имплементация механизма внимания**: На уровне графов StemGNN может внедрять механизмы внимания для более тонкой настройки на важные пространственно-временные взаимодействия.

Интеграция этих компонентов формирует сильный инструмент для обработки и анализа пространственно-временных данных, SteGNN эффективно может быть использована в самых разных областях, от метеорологии до городского планирования и управления транспортными потоками. SteGNN является отражением стремления к более точному моделированию сложных систем, где пространственная и временная динамика тесно связаны.
